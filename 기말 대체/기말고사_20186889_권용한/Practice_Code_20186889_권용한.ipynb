```python
data = pd.read_csv(filepath)

features = data.columns[:-1]
X_data = data[features]
y_data = data['answer']


fields = list(X_data.columns)
correlations_all = abs(X_data[fields].corrwith(data['answer']))

limit_1_3=0.1
limit_all=0.1
for i in range (0,len(correlations_all)):
    if correlations_1_3[i] < limit_1_3 or correlations_all[i]<limit_all:
        X_data=X_data.drop('f'+str(i+1),axis=1)

x_train,x_test,y_train,y_test=train_test_split(X_data,y_data,
                                               test_size=0.3,
                                               shuffle=True,
                                               random_state=42,
                                               stratify=y_data)
```

이후 전체 데이터(1,2,3 모두 고려)에서 correlation을 다시 체크해 최종 feature을 선정합니다.

## KNN

```python
# Mute the sklearn warning
import warnings
warnings.filterwarnings('ignore', module='sklearn')

#scaling = MinMaxScaler()
scaling = StandardScaler()
#scaling=MaxAbsScaler()

x_train = pd.DataFrame(scaling.fit_transform(x_train),columns=x_train.columns)
x_test = pd.DataFrame(scaling.fit_transform(x_test),columns=x_test.columns)
```

```python
# Function to calculate the % of values that were correctly predicted

def accuracy(real, predict):
    return sum(real == predict) / float(real.shape[0])
```

```python
# Sample code - kNN Classification
acc=0
selected_k=0
for k in range (1,51,2):
    model = KNeighborsClassifier(n_neighbors=k,p=1)#manhattan

    model = model.fit(x_train, y_train)

    y_train_pred=model.predict(x_train)    
    
    loo=LeaveOneOut()
    cv_N=10
    scores = cross_val_score(model, x_train, y_train, cv=cv_N)
    if acc<=np.mean(scores):
        acc=np.mean(scores)
        selected_k=k
model = KNeighborsClassifier(n_neighbors=selected_k,p=1)
model = model.fit(x_train, y_train)
y_test_pred = model.predict(x_test)
```

```python
print(accuracy(y_train, y_train_pred))
print(accuracy(y_test, y_test_pred))

cf_train=confusion_matrix(y_train,y_train_pred)
print(cf_train)

cf_test=confusion_matrix(y_test,y_test_pred)
print(cf_test)
```
